```python
import pandas as pd
import glob

# Get all CSV files in a folder
files = glob.glob("MergeFiles/*.csv")

# Read and combine
df = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)

print(df.head())
```

```python
df["info_count"] = df.notna().sum(axis=1)
df = df.sort_values("info_count", ascending=False)

df_best = df.drop_duplicates(subset="movie_id", keep="first")
df_best = df_best.drop(columns=["info_count"])

df_best.to_csv("final_output.csv", index=False)
```
