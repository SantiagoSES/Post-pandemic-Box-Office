```python
import pandas as pd

# Load the CSV
df = pd.read_csv("cleaned_fmbp_oneencoded.csv")

print(df['certificate'].unique())
```

```python
# Remove rows where "profit" is empty (NaN or blank)
df = df[df['certificate'] != '']

# Save back to CSV
df.to_csv("cleaned2_fmbp_oneencoded.csv", index=False)
```
