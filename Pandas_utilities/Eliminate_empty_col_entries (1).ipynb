```python
import pandas as pd

# Load the CSV
df = pd.read_csv("cleansed8_fmbp_oneencodedf.csv")

# Remove rows where "profit" is empty (NaN or blank)
df = df[df['budget'].notna() & (df['budget'].astype(str).str.strip() != "")]

#If we also don't want zeros.
df = df[df['budget'] >= 10000]

# Save back to CSV
df.to_csv("cleansed9_fmbp_oneencoded.csv", index=False)
```
