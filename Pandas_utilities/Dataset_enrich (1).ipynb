```python
import pandas as pd

df1 = pd.read_csv("cleansed11_fmbp_oneencoded.csv")
df2 = pd.read_csv("tn.movie_budgets.csv")

df2.rename(columns={"movie": "movie_name"}, inplace = True)

df1["movie_name"] = df1["movie_name"].astype(str)
df2["movie_name"] = df2["movie_name"].astype(str)

# Merge on common column
merged = pd.merge(df1, df2[['movie_name', 'worldwide_gross']], on="movie_name", how="left")

merged.to_csv("cleansed12_fmbp_oneencoded.csv", index=False)
```
