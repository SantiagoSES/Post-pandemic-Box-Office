```python
import pandas as pd

# Load the CSV
df = pd.read_csv("PostCovid2.csv")

#Remove empty spaces first
#df['genres'] = df['genres'].str.replace(' ', '')

# One-hot encode multi-value column
dummies = df["Certification"].str.get_dummies(sep=",")

# Join back to original dataframe
df = df.join(dummies)
df.to_csv("PostCovid3.csv", index=False)
```
